```python
import matplotlib.pyplot as plt
x=[-2,-1,1,2]
y=[-1,1,5,7]
plt.plot(x,y,marker='o',markersize=10)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
x1=[2,4,6,8,10]
y1=[1,2,3,4,5]
x2=[10,8,6,4,2]
y2=[1,2,3,4,5]
plt.plot(x1,y1,color='green',label="line1")
plt.plot(x2,y2,color='red',label="line2")
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x1=np.arange(1,12)
y1=np.arange(1,100,9)
x2=np.arange(12,1,-1)
y2=np.arange(1,100,9)
plt.plot(x1,y1,color='green',label="line1")
plt.plot(x2,y2,color='red',label="line2")
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.legend()
plt.show()
```
